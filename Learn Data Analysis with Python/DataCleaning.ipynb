```python
import pandas as pd
import numpy as np
```

```python
filepath = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df = pd.read_csv(filepath, names=headers) 
```

```python
df.replace("?", np.nan, inplace = True)
```

```python
for column in missing_data.columns.values.tolist():
    missing_data_series = missing_data[column].value_counts()
    #keys can only be True/False, true indicates missing value
    if(True in missing_data_series.keys()):     
        print("{}:{}".format( 
            column,
            missing_data_series[True]
            ))
    
```

```python
var = ["normalized-losses", "bore", "stroke", "horsepower", "peak-rpm"]

def replaceNaN_mean(df, name):
    mean = df[name].astype("float").mean(axis = 0)
    df[name].replace(np.NaN, mean, inplace=True)

def replaceNan_mode(df, name):
    mode = df[name].value_counts().idxmax()
    df[name].replace(np.NaN, mode, inplace=True)

for column in var:
    replaceNaN_mean(df, column)


replaceNan_mode(df,'num-of-doors')
```

```python
#Missing Information has been filled using mean/mode
missing_data = df.isnull()
for column in missing_data.columns.values.tolist():
    missing_data_series = missing_data[column].value_counts()
    #keys can only be True/False, true indicates missing value
    if(True in missing_data_series.keys()):     
        print("{}:{}".format( 
            column,
            missing_data_series[True]
            ))
    
```

```python
#price is dropped, as a model will be built to predict prices
df.dropna(subset=["price"], axis=0, inplace=True)
# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
```

<h3> Data Normalization </h3>
<p> Data types should be in a similiar ranges <p>

```python
#simple feature scaling
def norm_sfc(df, label):
    df[label] = df[label]/df[label].max()
    return  df

#min-max
def norm_minmax(df, label):
    df[label] = (df[label] - df[label].min())/(df[label].max() - df[label].min())
    return df
    
#z-score
def norm_zscore(df, label):
    df[label] = (df[label] - df[label].mean(axis = 0))/(df[label].std(axis = 0))
    return df
    
```

```python
df_norm_sfc = df.copy()
df_norm_sfc = norm_sfc(df_norm_sfc, "length")
df_norm_sfc = norm_sfc(df_norm_sfc, "width")

df_norm_minmax = df.copy()
df_norm_minmax = norm_minmax(df_norm_minmax, "length")
df_norm_minmax = norm_minmax(df_norm_minmax, "width")

df_norm_zscore = df.copy()
df_norm_zscore = norm_zscore(df_norm_zscore, "length")
df_norm_zscore = norm_zscore(df_norm_zscore, "width")
```

```python
comparison = [df_norm_sfc["length"], df_norm_minmax["length"], df_norm_zscore["length"],
              df_norm_sfc["width"], df_norm_minmax["width"], df_norm_zscore["width"]]
header_comparison = ["sfc:length", "min-max:length", "z-score:length",
                     "sfc:width", "min-max:width", "z-score:width"]
df_comparison = pd.DataFrame(columns = header_comparison)
df_comparison = pd.concat(comparison, axis=1, ignore_index=True)
df_comparison.columns = header_comparison
df_comparison.head(5)

```
